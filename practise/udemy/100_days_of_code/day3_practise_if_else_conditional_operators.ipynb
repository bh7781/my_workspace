# Nested if statements and elif statements
# & Multiple if statements in succession

```python
height = int(input('What is your height in cm?: '))
```

```python
bill = 0
if height >= 120:
    print('You can ride the rollercoaster')
    age = int(input('What is your age?: '))
    if age <= 12:
        bill = 5
        print("Child tickets are $5")
    elif age <= 18:
        bill = 7
        print("Youth tickets are $7")
    else:
        bill = 12
        print("Adult tickets are $12")

    wants_photo = input("Do you want to have a photo taken? Type Y for yes and N for no: ")
    if wants_photo.upper() == "Y":
        bill += 3
        print(f"Your bill is ${bill}.\nYou have opted for a photograph in your ride.")
    else:
        print(f"Your bill is ${bill}.")
else:
    print('Sorry, you have to grow taller before you can ride.')
```

# Pizza Order Practise

```python
print("Welcome to Python Pizza Deliveries!")

def get_valid_input(prompt, valid_options):
    while True:
        response = input(prompt)
        if response.upper() in valid_options:
            return response.upper()
        else:
            print(f"Invalid choice. You entered: '{response}'. Please enter one of: {', '.join(valid_options)}")

size = get_valid_input("What size pizza do you want? S, M, or L: ", ["S", "M", "L"])
pepperoni = get_valid_input("Do you want pepperoni on your pizza? Y or N: ", ["Y", "N"])
extra_cheese = get_valid_input("Do you want extra cheese? Y or N: ", ["Y", "N"])

price = {"S": 15, "M": 20, "L": 25}
pepperoni_price = {"S": 2, "M": 3, "L": 3}


bill = price[size]

if pepperoni == "Y":
    bill += pepperoni_price[size]
if extra_cheese == "Y":
    bill += 1

print(f"""
      -------Your order preview-------
      Pizza Size:   {size} ${price[size]}
      Pepperoni:    {pepperoni} ${pepperoni_price[size]}
      Extra cheese: {extra_cheese} ${1 if extra_cheese else 0}
      --------------------------------
      Your bill amount: ${bill}
""")
```
